# Loops

+ A programming structutre that implements iteration is called Loops.
+ Iteration : 2 types
  + Indefinite Iteration
  + Definite Iteration
  

## While Loop

+ we can execute a set of statements as long as a condition is true.
+ While loop falls under the category of indefinite iteration. 
+ Indefinite iteration means that the number of times the loop is executed isnâ€™t specified explicitly in advance.

## For Loop

+ For loops, in general, are used for sequential traversal. 
+ It falls under the category of definite iteration. 
+ Definite iterations means the number of repetitions is specified explicitly in advance. 

```python
i = 5
while i>1:
    print(i)
    i -=1 # i = i-1
```

```python
for i in range(5, 1, -1):
    print(i)
```

```python
fruits = ['apple', 'pineapple', 'banana'] #no index maintaining

while fruits:
    print(fruits.pop())
    print(fruits)
    
```

```python
fruits = ['apple', 'strawberry', 'pineapple', 'banana'] #based on index
j=0
for fruit in fruits:
    print(j)
    print(fruit)
    
    print(fruits)
    j+=1
```

## Loop control statements
### break statement

```python
#print fruits till we find 'strawberry' 
fruits = ['apple', 'strawberry', 'pineapple', 'banana']
i = 0
while i<len(fruits):
    if fruits[i]=="strawberry":
        break
    print(fruits[i])
    i+=1
```

```python
fruits = ['apple', 'strawberry', 'pineapple', 'banana']
for i in fruits:
#     if i == "strawberry":
#         break
    print(i)
        
```

### continue statement

```python
#print fruits other then 'strawberry' 
fruits = ['apple', 'strawberry', 'pineapple', 'banana']
i = 0
while i<len(fruits):
    if fruits[i]=="strawberry":
        i+=1
        continue
    print(fruits[i])
    i+=1
    
```

```python
fruits = ['apple', 'strawberry', 'pineapple', 'banana']
for fruit in fruits:
    if fruit == "strawberry":
        continue
    print(fruit)
```

### else statement with while and for loop

```python
#print fruits till we find 'strawberry' or if strawberry is not present than print "Not Found"
fruits = ['apple', 'strawberry', 'pineapple', 'banana']
i = 0
while i<len(fruits):
    if fruits[i]=="orange":
        break
    print(fruits[i])
    i+=1
else:
    print("Not Found")
```

```python
fruits = ['apple', 'strawberry', 'pineapple', 'banana']
for fruit in fruits:
    if fruit == "guava":
        break
    print(fruit)
else:
    print("Not Found")
```

### Nested loops

```python
numbers = [1,2,3]
chars = ['a','b','c']
i = 0

while i< len(numbers):
    j = 0
    while j<len(chars):
        print(numbers[i],chars[j])
        j+=1
    i+=1
```

```python
numbers = [1, 2, 3]
chars = ["a", "b", "c"]

for x in numbers:
    for y in chars:
        print(x, y)
```
